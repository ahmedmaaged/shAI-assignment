# 3. Data Cleaning

```python
#drop null columns
df.drop(columns=['Notes', 'Status'], inplace=True)
df.isnull().sum()
```

```python
#drop null base
df.dropna(subset=["BasePay"], axis=0, inplace=True)
df.isnull().sum()
```

```python
df1 = df[(df['Benefits']==0)]
df1
```

```python
#when benefits = zero then totalpaybenefits should equal the totalpay 
#if not >> this data is corupted (drop)
df1 = df[(df['Benefits'].isna()) & (df['TotalPayBenefits']==df['TotalPay'])]
df1
```

```python
#all data is good so replace nan with zero
df['Benefits'] = df['Benefits'].fillna(0)
df.isnull().sum()
```

# 4. Basic Data Visualization

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=5, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df1['TotalPay'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
df1['JobTitle'].value_counts()
```

```python
#extracting nimber of different JobTitle to a list
b = list(map(list, df1["JobTitle"].value_counts().items()))
#show an example of the list
a = b[1:10]
a
```

```python
#selecting the JobTitle with low number of data
b = list(map(list, df1["JobTitle"].value_counts().items()))
f = [b[m][0] for m in range(len(b)) if b[m][1] < 1000 ]
a = len(f)
a
```

```python
#Dropping them
df1 = df1[~df1['JobTitle'].isin(f)]
df1['JobTitle'].value_counts()
```

```python
print(df1['JobTitle'].unique())
```

```python
#Grouping them
job_title_to_department = {
    'REGISTERED NURSE': 'Health',
    'SPECIAL NURSE': 'Health',
    'TRANSIT OPERATOR': 'Transportation',
    'Registered Nurse': 'Health',
    'Firefighter': 'Emergency Services',
    'Attorney (Civil/Criminal)': 'Legal',
    'Police Officer 3': 'Law Enforcement',
    'Police Officer 2': 'Law Enforcement',
    'Police Officer': 'Law Enforcement',
    'Deputy Sheriff': 'Law Enforcement',
    'Transit Operator': 'Transportation',
    'Special Nurse': 'Health',
    'General Laborer': 'General',
    'Patient Care Assistant': 'Health',
    'Porter': 'General',
    'Custodian': 'General',
    'Public Service Trainee': 'General',
    'Public Svc Aide-Public Works': 'Public Works',
    'Recreation Leader': 'Recreation'
}

# Create a new column 'Department' based on the mapping
df1['Department'] = df1['JobTitle'].map(job_title_to_department)

df1['Department'].value_counts()
```

```python
department_counts = df1['Department'].value_counts()
plt.figure(figsize=(15, 8))
plt.bar(department_counts.index, department_counts, color='lightcoral')
plt.title('Proportion of Employees in Different Departments')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

# 5. Group Analysis

```python
# Group the data by 'Department' and calculate average salary for each department
average_salary_by_department = df1.groupby('Department')['TotalPay'].mean().reset_index()

# Plot the average salaries across different departments using a bar chart
plt.figure(figsize=(14, 6))
plt.bar(average_salary_by_department['Department'], average_salary_by_department['TotalPay'], color='skyblue')
plt.title('Average Salary Across Different Departments')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.grid(axis='y')
plt.show()
```

```python
# Group the data by 'Department' and calculate average benefits for each department
average_benefits_by_department = df1.groupby('Department')['Benefits'].mean().reset_index()

# Plot the average salaries across different departments using a bar chart
plt.figure(figsize=(14, 6))
plt.bar(average_benefits_by_department['Department'], average_benefits_by_department['Benefits'], color='skyblue')
plt.title('Average Benefits Across Different Departments')
plt.xlabel('Department')
plt.ylabel('Average Benefits')
plt.grid(axis='y')
plt.show()
```

```python
# Group the data by 'Department' and calculate average overtimepay for each department
average_overtime_by_department = df1.groupby('Department')['OvertimePay'].mean().reset_index()

# Plot the average salaries across different departments using a bar chart
plt.figure(figsize=(14, 6))
plt.bar(average_overtime_by_department['Department'], average_overtime_by_department['OvertimePay'], color='skyblue')
plt.title('Average Overtime Pay Across Different Departments')
plt.xlabel('Department')
plt.ylabel('Average Overtime Pay')
plt.grid(axis='y')
plt.show()
```
